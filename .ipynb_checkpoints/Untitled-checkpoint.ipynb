```python
# --- Heatmap корреляций ---
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")

# Кодирование категорий
data_encoded = data.copy()
for col in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data_encoded[col] = le.fit_transform(data_encoded[col])

# Матрица корреляций
corr_matrix = data_encoded.corr().round(2)

# Отбор значимых корреляций
significant_corrs = corr_matrix[
    (corr_matrix.abs() > 0.3) & (corr_matrix != 1.0)
].dropna(how='all').dropna(how='all', axis=1)

# Вывод таблицы
mask = significant_corrs.isna()
significant_corrs.style \
    .background_gradient(cmap='coolwarm', axis=None) \
    .format(precision=2) \
    .apply(lambda x: ['background-color: white' if m else '' for m in mask.loc[x.name]], axis=1)
```

```python
# --- Bloxplot ---
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")
data = data.drop(columns=['Row_id'], errors='ignore')
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
prepared_data = data.select_dtypes(include=['int64', 'float64']).columns

for col in prepared_data:
    plt.figure(figsize=(4, 2))
    sns.boxplot(x=data[col])
    plt.xlabel(col)
    plt.grid(True)
    plt.show()

```

```python
# --- StandartScaler ---
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("data.csv")

# Удалить ненужные столбцы
data = data.drop(columns=['Row_id', 'SeniorCitizen'], errors='ignore')

# Тотал чарджи - obj, меняет на float
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')

numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Возвращаем в DataFrame
scaled_df = pd.DataFrame(StandardScaler().fit_transform(numeric_data), columns=numeric_data.columns)

print(scaled_df)
```
